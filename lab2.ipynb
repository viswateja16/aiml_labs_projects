```python
def split(data, label, percent):

    rnd = rng.random(len(label))

    train_data = data[rnd < percent]
    train_label = label[rnd < percent]

    test_data = data[rnd >= percent]
    test_label = label[rnd >= percent]

    return train_data, train_label, test_data, test_label
```
